```python
import pandas as pd
import numpy as np
from sklearn.tree import DecisionTreeClassifier, export_text
from sklearn.preprocessing import LabelEncoder

# Cargar y preparar datos
df = pd.read_csv('alumnos_ujed_actualizado.csv')

le = LabelEncoder()
df['Sexo_N'] = le.fit_transform(df['Sexo'])        
df['Trabaja_N'] = le.fit_transform(df['Trabaja'])  

df['Reprobo_Estructura_Datos'] = [0, 1, 1, 0, 1, 0, 0, 0, 1, 0, 0, 1, 0, 0, 1, 1]

X = df[['Matematicas', 'Contabilidad_I', 'Habilidades_Pensamiento', 'Sexo_N', 'Trabaja_N']]
y = df['Reprobo_Estructura_Datos']

print("¡Datos listos y cargados en memoria!")
```

```python
# Entrenar el Árbol de Decisión
modelo = DecisionTreeClassifier(max_depth=3, random_state=42)
modelo.fit(X, y)

# Mostrar las reglas de forma interactiva
reglas = export_text(modelo, feature_names=list(X.columns))
print("--- REGLAS DE DECISIÓN DEL MODELO ---")
print(reglas)
```

```python
# AQUÍ PUEDES JUGAR: Cambia estos números cuantas veces quieras y corre solo esta celda
# Orden: [Matemáticas, Contabilidad, Habilidades, Sexo(0=F, 1=M), Trabaja(0=No, 1=Sí)]
valores_alumno = [[3.0, 8.0, 7.5, 0, 0]] 

nuevo_alumno = pd.DataFrame(valores_alumno, columns=list(X.columns))

# Predicción
prediccion = modelo.predict(nuevo_alumno)
probabilidad = modelo.predict_proba(nuevo_alumno)

print("--- RESULTADO DE LA EVALUACIÓN ---")
if prediccion[0] == 1:
    print("Resultado: ¡ALERTA DE RIESGO! (Reprobará)")
else:
    print("Resultado: Alumno a salvo (Aprobará)")

print(f"\nProbabilidad de aprobación: {probabilidad[0][0] * 100:.1f}%")
print(f"Probabilidad de reprobación: {probabilidad[0][1] * 100:.1f}%")
```
